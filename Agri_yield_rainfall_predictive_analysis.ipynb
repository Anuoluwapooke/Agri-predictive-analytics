# Predictive Analysis of Rainfall Impact on Crop Yield.




## Project Statement
This project analyzes how rainfall influences crop yield and identifies
risk periods that may impact agricultural supply chains.

#Problem:
Understanding how rainfall influences crop yield to support agricultural planning.
#Data:
Historical crop yield and rainfall data across multiple seasons.
#Analysis:
Exploratory analysis and visualizations show a positive correlation between rainfall and crop yield. Low-rainfall seasons are associated with reduced yields and increased production risk.
#Prediction:
If rainfall falls below the historical average, crop yield is likely to decrease.
#Recommendation:
Farmers should implement drip irrigation and water management strategies during low-rainfall seasons to stabilize yields and protect supply chain output.

# Data Visualization

```python
import matplotlib.pyplot as plt
```

Rainfall shows variability with a notable decline in 2020, followed by recovery in subsequent years.

```python
plt.plot(df["Year"], df["Rainfall_mm"])
plt.xlabel("Year")
plt.ylabel("Rainfall (mm)")
plt.title("Rainfall Trend Over Time")
plt.show()
```

Although yield does not follow rainfall perfectly year by year, higher rainfall years are associated with higher yields.

```python
plt.plot(df["Year"], df["Yield_tons"])
plt.xlabel("Year")
plt.ylabel("Yield (tons)")
plt.title("Crop Yield Trend Over Time")
plt.show()
```

While rainfall and yield trends over time are not perfectly aligned, the scatter plot indicates a positive correlation between rainfall and crop yield. This suggests that rainfall is a key predictive factor for yield, and below-average rainfall seasons present increased production risk.

```python
plt.scatter(df["Rainfall_mm"], df["Yield_tons"])
plt.xlabel("Rainfall (mm)")
plt.ylabel("Yield (tons)")
plt.title("Rainfall vs Crop Yield")
plt.show()
```

## Key Insights
- Crop yield increases in seasons with above-average rainfall
- Low rainfall seasons (e.g., 2020) show higher yield risk
- Rainfall and yield show a positive correlation

## Prediction & Recommendation
If rainfall falls below the historical average, crop yields are expected
to decline. Farmers should adopt drip irrigation strategies to mitigate
production risk and stabilize supply chain output.
